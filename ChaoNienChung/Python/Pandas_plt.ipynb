- 利用 matlibplot 畫出多支個股走勢圖
- 視覺化很重要！！！
- 光看數字是沒有感覺的，只有圖畫出來最清楚

```python
# 引入模組

import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
# 讀取多支個股資料
# Join 到同一個 DataFrame

Symbols = ["2330", "2317", "2454", "2881", "2308", "2382", "2412", "2882", "2891", "3711"]

start = datetime.date(2024, 1, 1)
end = datetime.date.today()
DateRange = pd.date_range(start, end)

df_Symbols = pd.DataFrame(index=DateRange)
df_Symbols.index = pd.to_datetime(df_Symbols.index)

for symbol in Symbols:

    df = pd.read_csv(f"C:\\__STOCK\\DATA\\TWSE\\{symbol}.csv", usecols=["Date", "Close"])
    df["Date"] = pd.to_datetime(df["Date"], format="%Y%m%d")
    df = df.set_index("Date")
    df = df.rename(columns={"Close":f"{symbol}"})

    df_Symbols = df_Symbols.join(df).dropna()


plt.style.use("bmh")

plt.subplot(4,1,1)
df_Symbols[Symbols[0]].plot(title=f"{Symbols[0]}.TW", fontsize=8, figsize=(25, 17))

plt.subplot(4,3,4)
df_Symbols[Symbols[1]].plot(title=f"{Symbols[1]}.TW", fontsize=8)

plt.subplot(4,3,5)
df_Symbols[Symbols[2]].plot(title=f"{Symbols[2]}.TW", fontsize=8)

plt.subplot(4,3,6)
df_Symbols[Symbols[3]].plot(title=f"{Symbols[3]}.TW", fontsize=8)

plt.subplot(4,3,7)
df_Symbols[Symbols[4]].plot(title=f"{Symbols[4]}.TW", fontsize=8)

plt.subplot(4,3,8)
df_Symbols[Symbols[5]].plot(title=f"{Symbols[5]}.TW", fontsize=8)

plt.subplot(4,3,9)
df_Symbols[Symbols[6]].plot(title=f"{Symbols[6]}.TW", fontsize=8)

plt.subplot(4,3,10)
df_Symbols[Symbols[7]].plot(title=f"{Symbols[7]}.TW", fontsize=8)

plt.subplot(4,3,11)
df_Symbols[Symbols[8]].plot(title=f"{Symbols[8]}.TW", fontsize=8)

plt.subplot(4,3,12)
df_Symbols[Symbols[9]].plot(title=f"{Symbols[9]}.TW", fontsize=8)
```
